# Weight for it

Find weights for two dice so that the probability of rolling 2, 3, ..., 12 is the same.

Or, prove that not such weighting exists.

```python
import numpy as np
import pandas as pd

import plotly.graph_objects as go
```

```python
"""

    define sample space of the two dice

"""

d1 = np.arange( 1, 7, 1 )
d2 = np.arange( 1, 7, 1 )

print(f'd1: {d1}')
print(f'd2: {d2}')
```

```python
def roll_simulation(sim, d1, d2):
    
    if sim > 0:
        
        result = []
        
        for i in range(sim):
            
            r1 = np.random.choice(d1)
            r2 = np.random.choice(d2)
        
            r = r1 + r2
            
            result.append( r )
            
    else:
        
        print('the input has to be greater than zero')
        
    return result
        
rs_1 = roll_simulation(1000, d1, d2)
rs_2 = roll_simulation(100000, d1, d2)

fig = go.Figure()

fig.add_trace(go.Histogram(x=rs_1, name='rs_1', histnorm='probability density'))
fig.add_trace(go.Histogram(x=rs_2, name='rs_2', histnorm='probability density'))

# Overlay both histograms
fig.update_layout(barmode='overlay')
# Reduce opacity to see both histograms
fig.update_traces(opacity=0.75)
fig.show()
```
